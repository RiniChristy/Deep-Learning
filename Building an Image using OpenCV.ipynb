##**Creating an Image**
The python libraries cv2 needed to be imported before reading or displaying an image.

Methods like cv2.rectangle, cv2.line, cv2.circle, cv2.ellipse etc can be used to build an image like house or bus.

`cv2.waitKey()` is a keyboard binding function. Its argument is the time in milliseconds. The function waits for specified milliseconds for any keyboard event. If you press any key in that time, the program continues. If 0 is passed, it waits indefinitely for a key stroke. It can also be set to detect specific keystrokes like the Q key or the ESC key on the keyboard. 

`cv2.destroyAllWindows()` simply destroys all the windows we created. If you want to destroy any specific window, use the function `cv2.destroyWindow()` where you pass the exact window name as the argument.

```python
import cv2
w= cv2.waitKey(0)
if w == 27 or w == ord('q'):
    cv2.destroyAllWindows()

import numpy as np
pic = np.ones((1000,1000,3), dtype = 'uint8')
```

```python
#Door & Door Frame
#Draw a rectangle with white line borders of thickness of 3 px
#Format is cv2.rectangle(image, start_point, end_point, color, thickness)
#Start coordinate, here (400, 500) represents the top left corner of rectangle
#Ending coordinate, here (600,800) represents the bottom right corner of rectangle
#White color in BGR
rect = cv2.rectangle(pic,(400, 500), (600,800), (255,255,255), 3 )
rect = cv2.rectangle(pic,(410, 510), (590,800), (255,255,255), 3 )
```

##**Displaying an image**
* Display the image using plt.imshow() function

The function `cv2.imshow()` is used to display an image in a window. This function takes 2 arguments. The first argument is the window name (Specify unique window names to display multiple images) and the second argument is the image to be displayed.

All the OpenCV interactive functions won't work in any Jupyter/Colab notebook environment. The function cv2.imshow() opens a window displaying the image and therefore cannot be used in a Colab notebook. So, cv2.imshow() won't be able to use here. Matplotlib is used instead to display image in the notebook. To use Matplotlib, import the matplotlib library.

To hide the tick values on x and y axis, the function plt.axis("off") can be used.

The function plt.imshow() is used to specify the image to be displayed and the function plt.show() is used to display it.

The displayed image won't be as expected. This happens because OpenCV represents RGB images as multi-dimensional NumPy array but in the reverse order. This means that images are actually represented in BGR order rather than RGB. This can be corrected by converting BGR to RGB.


```python
#Displaying the image with format cv2.imshow(window_name, image) 
#cv2.imshow('House_image', rect)
#w= cv2.waitKey(0)
#press any key to close the windows
cv2.destroyAllWindows()
from matplotlib import pyplot as plt
plt.figure(figsize=(8,8), dpi = 100)
plt.axis("off")
#plt.imshow(cv2.cvtColor(rect, cv2.COLOR_BGR2RGB))
plt.imshow(rect)
plt.show()
```

##**Image Resizing**
###**Resize by specifying new width and height**

```python
plt.figure(figsize=(8,8), dpi = 100)
resized = cv2.resize(rect, (2000,2000), interpolation= cv2.INTER_LINEAR)
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
resized = cv2.resize(rect, (5000,5000), interpolation= cv2.INTER_LINEAR)
plt.figure(figsize=(8,8), dpi = 100)
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
#RGB Format
plt.figure(figsize=(20,10), dpi = 100)
plt.imshow(resized)
plt.show()
```

```python
plt.figure(figsize=(20,10), dpi = 100)
plt.subplot(131); plt.imshow(cv2.cvtColor(rect, cv2.COLOR_BGR2RGB)); plt.title("Original Image")
plt.subplot(132); plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)); plt.title("Scaled Up Image in BGR format")
plt.subplot(133); plt.imshow(resized); plt.title("Scaled Up Image in RGB format")
plt.show()
```

###**Resize by Specifying the Scaling Factor**

```python
#scaling Up factor
scale_Up_X = 1.5
scale_Up_Y = 1.5

#scaling Down factor 
scale_Down = 0.6

scaledDown = cv2.resize(rect, None, fx= scale_Down, fy= scale_Down, interpolation= cv2.INTER_LINEAR)

scaledUp = cv2.resize(rect, None, fx= scale_Up_X, fy= scale_Up_Y, interpolation= cv2.INTER_LINEAR)
# We can also use the following syntax for displaying image
plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(rect[...,::-1])
plt.title("Original Image, size = {}".format(rect.shape[:2]))
plt.subplot(132);plt.imshow(scaledDown[...,::-1])
plt.title("Scaled Down Image, size = {}".format(scaledDown.shape[:2]))
plt.subplot(133);plt.imshow(scaledUp[...,::-1])
plt.title("Scaled Up Image, size = {}".format(scaledUp.shape[:2]))
plt.show()
```

### Interpolation
* Different Interpolation methods are used for different functions   
 * **cv2.INTER_AREA** is best used for **shrinking** of Images because it resamples the image using pixel area relation as it gives more free results 
 * **cv2.INTER_CUBIC** (slow) is  best used for **zooming** of the Image , it uses a bicubic interpolation over **4x4 pixel** neighborhood

 * **cv2.INTER_LINEAR**  uses a bilinear interpolation and is  used for all resizing purposes, hence it is used as **default** in many resizing functions .

 * If you prefer speed over quality use a **Nearest Neighbor method** but if you want to preserve the quality after downsampling then **interpolation** is a better option but it also depends on image content
